### Задание 4
### 1) Решение системы:
#### y = x**2 -1
#### exp(x) + x(1-y) = 1

```python
from scipy.optimize import fsolve
```

```python
x = np.linspace(-3, 5 , 201)
```

```python
# Сводим задачу к одной переменной
plt.plot(x, np.exp(x) + x*(2-x**2) - 1)
plt.plot(x, x**2 - 1,'--')
plt.plot([-3, 5], [0, 0])
plt.grid()
```

```python
def equation(x):
    return np.exp(x) + x*(2-x**2) - 1
```

```python
x= fsolve(equation, -2)[0], fsolve(equation, 0)[0], fsolve(equation, 2)[0], fsolve(equation,5)[0]
```

```python
for ind, val in enumerate(x):
    print(f'x{ind}={val} y{ind}={val**2-1}')

y = [x**2-1 for x in x]
plt.scatter(x, y)
plt.grid()
```

### 2) Решение системы:
#### y = x**2 -1
#### exp(x) + x(1-y) > 1

```python
# Собственно, решение задачи следует из графика и корней, найденных в предыдущей задаче:
```

```python
x = np.linspace(-3, 5 , 401)
y = [x**2-1 if np.exp(x) + x*(2-x**2) - 1 > 0 else np.nan for x in x]
plt.plot(x, y, 'o')
plt.plot([-3, 5], [0, 0])
plt.grid()
```
