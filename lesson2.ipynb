```python
def vectorlen(vector):
    vl=0
    for v in vector:
        vl +=v**2
    return np.sqrt(vl)
```

#### Задание 3

```python
from matplotlib import pyplot as plt
import matplotlib.ticker as ticker

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
y=[[],[],[]]
```

```python
x = np.linspace(-10,10,1000)
tt_circle = np.array(3*3-x**2) # окружность радиуса 3
tt_elips = np.array(1-x**2/5/5) 
tt_hyper = np.array(-1+x**2/1/1)

y[0] = [np.sqrt(t) if t>=0 else np.nan for t in tt_circle]
y[1] = [3*np.sqrt(t) if t>=0 else np.nan for t in tt_elips] # элипс с полуосями 5 и 3
y[2] = [1*np.sqrt(t) if t>=0 else np.nan for t in tt_hyper] # гипербола с параметрами 1 и 1
```

```python
fig, ax = plt.subplots(1, 3,figsize=(12,6))
for i in range(3):
    ax[i].xaxis.set_major_locator(ticker.MultipleLocator(1))
    ax[i].yaxis.set_major_locator(ticker.MultipleLocator(1))
    ax[i].plot(x, y[i], '--', color='blue')
    ax[i].plot(x, [-t for t in y[i]],'--', color='blue')
    ax[i].set_xlim(-6,6)
    ax[i].set_ylim(-4,4)
    ax[i].set_aspect("equal")
    ax[i].grid()
```

```python
X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
```

```python
X, Y = np.meshgrid(X, Y)
```

```python
Z_el = 2.25*(1 - X**2/9 - Y**2/16) # заготовка под эллипсоид
Z_hyp = X**2 + Y**2 + 16 # заготовка под двух полостный гиперболоид
```

```python
# вычисление квадратного корня от Z для элипсоида 
a = Z_el>=0
Z_el1 = Z_el.copy()
for i in range(Z_el.shape[0]):
    for j in range(Z_el.shape[1]):
        if a[i,j]:
            Z_el[i,j] = np.sqrt(Z_el[i,j])
            Z_el1[i,j] = -np.sqrt(Z_el1[i,j])
        else:
            Z_el[i,j] = np.nan
            Z_el1[i,j] = np.nan
```

```python
# вычисление квадратного корня от Z для гиперболоида
a = Z_hyp>=0
Z_hyp1 = Z_hyp.copy()
for i in range(Z_hyp.shape[0]):
    for j in range(Z_hyp.shape[1]):
        if a[i,j]:
            Z_hyp[i,j] = np.sqrt(Z_hyp[i,j])
            Z_hyp1[i,j] = -np.sqrt(Z_hyp1[i,j])
        else:
            Z_hyp[i,j] = np.nan
            Z_hyp1[i,j] = np.nan
```

```python
fig = figure()
ax = Axes3D(fig)

ax.plot_surface(X, Y, Z_hyp1, color='red') 

ax.plot_surface(X, Y, Z_el, color='green')
ax.plot_surface(X, Y, Z_el1, color='green')

ax.plot_surface(X, Y, Z_hyp, color='red')

fig.show()
```
